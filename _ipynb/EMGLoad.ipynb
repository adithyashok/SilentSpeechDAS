```python
import sys
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import h5py

#Function demonstrating data use without bsnb
#col[-1] gives the last channel
def pltEMG_mV(path, fname, channels):
  x=np.loadtxt(path + fname)
  i = 1
  while(i <= channels):
    plt.figure()
    plt.plot((x[0:23000,-i]/(2**10)-0.5)*3.3/1009*1000,'k')
    plt.ylim([-1.63,1.63])
    plt.ylabel('mV')
    plt.xlabel('t (ms)')
    plt.title(fname + ':EMG_Channel_' + str(channels - i + 1))
    i = i + 1


sample = 'Sample.txt'

x=np.loadtxt('/content/gdrive/MyDrive/__SPRING 2022/SNOD/SilentSpeechDAS/_SampleData/' + sample)

#f = h5py.File('/content/gdrive/MyDrive/__SPRING 2022/SNOD/SilentSpeechDAS/_SampleData/SampleEMG.h5')["20:16:02:26:60:88"]
#label = f["raw"]["channel_1"].attrs["label"]

#figure(figsize=(8,3))
plt.plot((x[18000:23000,-1]/(2**10)-0.5)*3.3/1009*1000,'k')
plt.ylim([-1.63,1.63])
plt.ylabel('mV')
plt.xlabel('t (ms)')
plt.savefig(sample + '.png',dpi=300) #,bbox_inches='tight')
```
